```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
t = sns.load_dataset("titanic")
```

```python
s = sns.load_dataset("flights")
```

```python
i = sns.load_dataset("fmri")
```

```python
plt.figure(figsize=(25,5))

plt.subplot(1,3,1)
sns.barplot(data=t,x="class",y="survived",hue="sex")


plt.subplot(1,3,2)
sns.boxplot(data=s ,x="year",y="passengers")


plt.subplot(1,3,3)
sns.scatterplot(data=i,x="timepoint", y="signal")
plt.title("fmri")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
```
